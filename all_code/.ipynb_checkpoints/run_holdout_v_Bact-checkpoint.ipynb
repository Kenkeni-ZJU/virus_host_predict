### Alter get_class_lists this for holdout virus group
use training with holdout data use fp.get_class_losts

```python
def get_class_lists_ho(subset,viruses,hosts):
    
    (label,label_tax,pool,pool_tax,balt,v_holdout) = subset
    data_lists = {'training':[[],[]],'test':[[],[]]}
    # Get a list of all the viruses in the labelled class and the rest of the pool
    
    
    baltimore = 'NA'  # all slasses (except satallites)
   
    for v, vd in viruses.items() :
        if  baltimore in vd['class']: 
            
            if  vd['family']!= v_holdout:
                ds_name ='training'
            else:
                ds_name = 'test'
            for h in vd['hosts']:
                if hosts[h][pool_tax.lower()] == pool:
                    if hosts[h][label_tax.lower()] ==label:
                        
                        data_lists[ds_name][0].append ((v,label))
                    else:
                        data_lists[ds_name][1].append ((v,'Other'))
                
           
   # print (f'training - positive {len(data_lists["training"][0])} negative {len(data_lists["training"][1])})') 
   # print (f'test - positive {len(data_lists["test"][0])} negative {len(data_lists["test"][1])}')

    # Get a random sample for each class of size n , the size of smallest class                    
   
   
 # Keep all the viruses in the holdout set as test data. they will not be same size   
    datasets = {'training':{},'test':{}}
    nmax = {'training':400,'test':50}
    for k,data_list in data_lists.items():
        datas =[]
        n = min (len(data_list[0]),len(data_list[1]),nmax[k])

        #print (f'number of v {k} {n}')
        #n = 6 #test
        # randomly choose n viruses from both classes
        for clss in data_list: 
            sample  = random.sample(range(len(clss)), n)
            for s in sample:
                datas.append (clss[s])
            
       # shuffle together
        random.shuffle(datas)
        for v,l in datas:
            datasets [k][v]= {'label':l, 'refseqs':viruses[v]['refseqs']}
    ntrain = (len(datasets['training']))
    ntest = (len(datasets['test']))
    #print(f' ho training v {ntrain} , test v {ntest}')
    return datasets,ntrain,ntest
```

### Adapt get_class_list to give infomation about number of holdout viruses in test set

```python
def get_class_lists(subset,viruses,hosts):
# returns two list of viruses one for each  positive class and negative class    
    (label,label_tax,pool,pool_tax,baltimore,v_ho) = subset
    
    if baltimore =='all':
        baltimore = 'NA'  # all classes  contain either 'DNA' or 'RNA'
    
    pos_neglists =  [[],[]]
    # Get 2 lists of all the viruses in the labelled class and the rest of the pool

    for v, vd in viruses.items() :
        if baltimore in vd['class']:          
            for h in vd['hosts']:
                if hosts[h][pool_tax.lower()] == pool:
                    if hosts[h][label_tax.lower()] ==label:
                        pos_neglists[0].append ((v,label))
                    else:
                         pos_neglists[1].append((v,'Other'))
    #print('all pos and neg v', len(pos_neglists[0]),len(pos_neglists[1]))                    
    # Get a random sample for each class of size n , the size of smallest class 
    #restrict to 400
    n = min(len(pos_neglists[0]),len(pos_neglists[1]),400)
   # n=6 # test
    random.seed(10)
    datas =  []
    for clss in pos_neglists:
        data = []
        sample  = random.sample(range(len(clss)), n)
        for s in sample:
            data.append (clss[s])
        datas.append(data)
    #print('pos and neg lists',len(datas[0]),len(datas[1]))   
    #split into training and test sets     
    split = 0.75 
    tntlist =  [[],[]]  #2 lists for training and test viruses
    
    for viruslist in datas:
        for i in range(n):
            if random.random() < split:
                tntlist[0].append(viruslist[i]) 
            else:
                tntlist[1].append(viruslist[i]) 
   
    datadicts = {'training':{},'test':{}}
    for i,k in enumerate(datadicts):
        random.shuffle(tntlist[i])

        for v,l in tntlist[i]:
            datadicts [k][v]= {'label':l, 'refseqs':viruses[v]['refseqs']}
       
        ntrain  =len(datadicts['training'])
        ntest = len(datadicts['test'])
        ho_test =[]
        for v in datadicts['test']:
            if viruses[v]['family']== v_ho:
                ho_test.append(v)
        n_ho_test = len (ho_test)
   # print (f'all- size of training and test set {ntrain},  {ntest}')
    
    return (datadicts,ntrain,n_ho_test)
    
    
    
```

```python
model =0
def test_dataset(subset,datasets,n): 
    (label,label_tax,pool,pool_tax,balt,v_holdout,t_g) = subset
    global model
    print(model)
    for feature_func in feature_functions:
        index = feature_functions.index(feature_func)
        filepath = filepaths[index]
        kmers = kmer_lists[index]
        symbol_dict = symbol_dicts[index]
        features = (feature_func.__name__).split('_')[0]
        for k in kmers:

            model += 1
        # get feature matrix , test prediction and write results to CSV file
            print (label,label_tax,feature_func.__name__ ,k,filepath)
            (X,X_test), v_index = feature_func (datasets,k,symbol_dict,filepath)
            y, y_test = fp.get_labels(label,v_index, datasets)
            print ( 'X_train,Y_train',np.shape(X), np.shape(y))
            results, pred_probs = fp.test_prediction2(X,y,X_test, y_test)
            # Add info for results
            results.update ({ 'N in class': n, 'Features':features, 'k': k})
            
            vs = []
            v_fam = []
            ys = []
            probs = []
            for  v,y,prob in zip(v_index[1],y_test,pred_probs):
                #print(v, hosts[viruses[v]['hosts'][0]][label_tax],y,prob)
                #print (viruses[v]['family'])
                if viruses[v]['family']== v_holdout:
                    vs.append(v)
                    v_fam.append(viruses[v]['family'])
                    ys.append(y)
                    probs.append(prob)
            print('length holdout data', len(vs),len(ys),len(probs))
            ss = (model,*subset,features,k)
            probs2csv(vs,v_fam,ys, probs,ss, dataset)
            results.update({'positive label':label ,'label tax group': label_tax,'pool label':pool,
                            'pool tax group':pool_tax, 'Baltimore':balt,
                            'virus holdout group':v_holdout,'training group':t_g,})
            print(results)
            results2CSV (results, results_file)
            
```

```python
def results2CSV(results, csvfile):
    fieldnames = ['positive label','label tax group','pool label','pool tax group',
                          'Baltimore','virus holdout group','training group', 'N in class' , 
                          'Features','k','AUC','accuracy', 'specificity','sensitivity', 'TN/FP/FN/TP' ]
    if os.path.isfile(csvfile):
        with open(csvfile, 'a') as csvfile:
            #fieldnames = ['positive label','label tax group','pool label','pool tax group','Baltimore','virus holdout group','training group', 'N in class' , 'Features','k','AUC','accuracy', 'specificity','sensitivity', 'TN/FP/FN/TP' ]
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
            writer.writerow(results) 
    else:
        with open(csvfile, 'a') as csvfile:
#             fieldnames = ['positive label','label tax group','pool label','pool tax group',
#                           'Baltimore','virus holdout group','training group', 'N in class' , 
#                           'Features','k','AUC','accuracy', 'specificity','sensitivity', 'TN/FP/FN/TP' ]
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
            writer.writeheader()
            writer.writerow(results)  
```

```python
def probs2csv (v_index, v_fam, y_test, pred_probs,subset, dataset):
    
    csvfile =  f'{dataset}_probs.csv'
    fieldnames = ['model','label','label_tax','pool','pool_tax','balt','v_holdout',
    'training group','features','k', 'Virus','virus family', 'true label', 'predict prob']
    
    if os.path.isfile(csvfile):
        with open(csvfile, 'a') as csvfile:
            writer = csv.writer(csvfile)
            for virus,f,y,y_pred in zip(v_index,v_fam,y_test, pred_probs):
                row = (*subset,virus,f,y,y_pred )
                writer.writerow(row) 
                
    else:
        with open(csvfile, 'w') as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow(fieldnames)
            for virus,f,y,y_pred in zip(v_index,v_fam, y_test, pred_probs):
                row = (*subset,virus,f,y,y_pred )
                writer.writerow(row) 
    
    
    return
```
